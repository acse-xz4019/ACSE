## Answer 1
The models are idealised and simplified reflections of reality, but they do not reflect all of the real world. They are approximations and assumptions, so they can never be said to be true. The name "approximations and assumptions" implies there are some errors. So that is why the statement says "all models are wrong".

However, sometimes, when people want to compute real-world problems with real-world data, they will find it can be expensive and computationally expensive. A simplification of a real-world complex problem then becomes a good way to save computing time and resources. Some models can do a great job implementing a great reflection of the real world with acceptable errors, such as the diffusion equation. So that is why the statement says "but some are useful".

"A model may get the right answer for the wrong reason." This statement implies that if people only perform a single calculation (e.g. for one numerical or physical parameter choice) without further checking, get an acceptable error, and say job done, there are probably bugs that appear so that the underlying model is wrong. Although the implementation is correct, the validation will fail.

The two statements are important because they tell people how to treat the models correctly. People should not undoubtedly believe in models, but sometimes, models are great tools for solving real-life problems. People should focus more on the underlying concept that forms the models instead of the models themselves.

As "model" developers, we can use **verification** and **validation** to mitigate the issues that these two statements may imply. 

**Verification** refers to the process of checking that the computational code or program accurately implements the intended mathematical model and makes sure that the code works well without bugs and as expected. The verification can be divided into
- code verification 

- solution verification

**Code verification** aims to find and eliminate bugs. A code verification testing hierarchy can be
1. Sanity tests - does it look correct
2. Trend/conservation tests - check that if an expected trend is present or if conservation laws can be checked
3. Benchmark test - check if the code can recreate an accepted community benchmark value
4. Comparison tests - cursory or detailed side-by-side comparison against another code
5. Against exact solutions

Convergence testing is an important technique in code verification. A convergence analysis often shows even the smallest, subtlest bugs unclear for a single run.

After ensuring that the codes work well and matches the idea of design (code validation), the next step is to check the actual answer or solution provided by the simulation to us. This verification is called **solution validation**. Solution verification aims to mitigate user errors and numerical errors.

Sensitivity testing is an important technique in solution verification to mitigate user errors, and it can test if the model is robust. Sensitivity testing checks whether the outputs are insensitive to users' choices.

**Validation** is about ensuring that the mathematical model itself is a true representation of the real-world physical situation it is meant to simulate. It includes comparing the model's predictions with actual experimental data to see if it can reliably reproduce the behaviour observed in the real world. **The responsibility of the model validation** is to identify the risk of the testing model; this includes the integrity of the assumption, the resources the data comes from, even if the model is robust.

Within the validation area, **model calibration** is vital; it aims to ensure that the distribution of results predicted from the model is consistent with that of the actual data. Usually, it is completed during the post-processing. Models are all assumptions; after the assumption is built, it is necessary to evaluate the quality of assumptions to make sure it is as accurate as possible. After the learning process, it is necessary to evaluate the quality of assumptions as accurately as possible. (Farrell, P.E. et al. (2011), Maitra, S. (2021))

Now it is better to raise an example about build a model to fit a function $f(x)=1.8 x +3$ to show the concepts of code verification, solution verification, model validation and model calibration.

```python
def exact_func(x):
    return 1.8 * x + 3

xi = np.array([-11.4, -8.9, -6.5, -3.2, -1.4, 2.2, 5.6, 8.0, ])

# add noise to simulate the real-life situation
yi = exact_func(xi) + np.random.normal(loc=0, scale=0.2, size=(np.shape(xi)))

# build the model
coeffis = np.polyfit(xi, yi, 1)
model = np.poly1d(coeffis)
x = np.linspace(-12, 9, 100)

# plot
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121)

ax1.plot(xi, yi, "o", label="raw data")
ax1.plot(x, model(x), "r", label="Model built")
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel('$f(x)$', fontsize=16)
ax1.set_title('Raw Data Plot', fontsize=16)
ax1.grid(True)
ax1.legend(loc='best', fontsize=8);

ax2 = fig.add_subplot(122)

# range of values we want to test
x2 = np.linspace(-5,5,10)

ax2.plot(x2, model(x2), 'b.-', label='our code')
ax2.plot(x2, exact_func(x2), 'r.-', label='"truth"')
ax2.set_xlabel('$x$', fontsize = 14)
ax2.set_ylabel('$f(x)$', fontsize = 14)
ax2.set_title('Code verfication (multiple runs/trend)', 
              fontsize = 16)
ax2.grid(True)
ax2.legend(loc='best', fontsize=14)
```

Consider a function of one independent variable: $f(x)$.

<br>

The first-order forward difference approximation to the derivative of $f$ at location $x$ is given by

$$ f'(x)\approx \frac{f(x+\Delta x)-f(x)}{\Delta x} $$

The second-order central difference approximation to the derivative of $f$ at location $x$ is given by 

$$ f'(x)\approx \frac{f(x+\Delta x)-f(x - \Delta x)}{2 \Delta x} $$

The fourth-order central difference approximation to the derivative of $f$ at location $x$ is given by

$$ f'(x)\approx \frac{-f(x + 2\Delta x) + 8 f(x+\Delta x)- 8 f(x - \Delta x) + f(x - 2\Delta x)}{12 \Delta x} $$

## Answer 

### accuracy of first-order forward difference approximation

From Taylor series got

$$ f({x+\Delta x}) = f(x) + \Delta xf'(x) + \frac{\Delta x^2}{2!}f''(x) + \frac{\Delta x^3}{3!}f'''(x) + \mathcal{O}(\Delta x^4),$$

$$ \Delta xf'(x) = f({x+\Delta x})- f(x)  - (\frac{\Delta x^2}{2!}f''(x) + \frac{\Delta x^3}{3!}f'''(x) + \mathcal{O}(\Delta x^4)),$$

$$ \Delta xf'(x) = f({x+\Delta x})- f(x)  - (\mathcal{O}(\Delta x^2)),$$

$$ f'(x) = \frac{f({x+\Delta x})- f(x)  - (\mathcal{O}(\Delta x^2))}{\Delta x},$$


The first order forward difference approxiamtion is given that

$$ f'(x)\approx \frac{f(x+\Delta x)-f(x)}{\Delta x}, $$

The local trunction error is

$$ \frac{\mathcal{O}(\Delta x^2)}{\Delta x} $$

$$ = \frac{\frac{\Delta x^2}{2!}f''(x) + \mathcal{O}(\Delta x^3))}{\Delta x} $$

$$ = \frac{\Delta x}{2!}f''(x) + \frac{\mathcal{O}(\Delta x^3))}{\Delta x} $$

The term $\frac{\Delta x}{2!}f''(x)$ shows the major error is proportion to $\Delta x$, making it a first-order approximation.


### accuracy of second-order central difference approximation
From Taylor series got

$$ f({x+\Delta x}) = f(x) + \Delta xf'(x) + \frac{\Delta x^2}{2!}f''(x) + \frac{\Delta x^3}{3!}f'''(x) + \mathcal{O}(\Delta x^4),$$

and

$$ f({x-\Delta x}) = f(x) - \Delta xf'(x) + \frac{\Delta x^2}{2!}f''(x) - \frac{\Delta x^3}{3!}f'''(x) + \mathcal{O}(\Delta x^4),$$

Hence,

$$ f({x+\Delta x}) - f({x-\Delta x}) = 2\Delta xf'(x) + \frac{\Delta x^3}{3}f'''(x) + ...,$$

$$ 2\Delta xf'(x) = f({x+\Delta x}) - f({x-\Delta x}) - \frac{\Delta x^3}{3}f'''(x) + ...,$$

$$ f'(x) = \frac{f({x+\Delta x}) - f({x-\Delta x}) - \frac{\Delta x^3}{3}f'''(x) + ...}{2\Delta x},$$

The second-order central difference approximation is given that

$$ f'(x)\approx \frac{f(x+\Delta x)-f(x - \Delta x)}{2 \Delta x} ,$$

The local truncation error is 

$$ \frac{- \frac{\Delta x^3}{3}f'''(x) + ...}{2\Delta x} $$

$$ = - \frac{\Delta x^2}{6}f'''(x) + ... $$

The term $- \frac{\Delta x^2}{6}f'''(x)$ shows the major error is proportion to $\Delta x^2$, making it a second-order approximation.

### accuracy of fourth-order central difference approximation
From Taylor series got

$$ f({x+\Delta x}) = f(x) + \Delta xf'(x) + \frac{\Delta x^2}{2!}f''(x) + \frac{\Delta x^3}{3!}f'''(x) + \frac{\Delta x^4}{4!}f''''(x) + \frac{\Delta x^5}{5!}f'''''(x) + \mathcal{O}(\Delta x^6),$$

and

$$ f({x-\Delta x}) = f(x) - \Delta xf'(x) + \frac{\Delta x^2}{2!}f''(x) - \frac{\Delta x^3}{3!}f'''(x) + \frac{\Delta x^4}{4!}f''''(x) - \frac{\Delta x^5}{5!}f'''''(x) + \mathcal{O}(\Delta x^6),$$

and

$$ f({x+2\Delta x}) = f(x) + 2\Delta xf'(x) + \frac{(2\Delta x)^2}{2!}f''(x) + \frac{(2\Delta x)^3}{3!}f'''(x) + \frac{(2\Delta x)^4}{4!}f''''(x) + \frac{(2\Delta x)^5}{5!}f'''''(x) + \mathcal{O}((2\Delta x)^6),$$

and

$$ f({x-2\Delta x}) = f(x) - 2\Delta xf'(x) + \frac{(2\Delta x)^2}{2!}f''(x) - \frac{(2\Delta x)^3}{3!}f'''(x) + \frac{(2\Delta x)^4}{4!}f''''(x) - \frac{(2\Delta x)^5}{5!}f'''''(x) + \mathcal{O}((2\Delta x)^6),$$

Hence,

$$ f({x+\Delta x}) - f({x-\Delta x}) = 2\Delta xf'(x) + \frac{\Delta x^3}{3}f'''(x) + \frac{\Delta x^5}{60}f'''''(x) + ...,$$

$$ f({x+2\Delta x}) - f({x-2\Delta x}) = 4\Delta xf'(x) + \frac{8\Delta x^3}{3}f'''(x) + \frac{8\Delta x^5}{15}f'''''(x) + ...,$$

Then,

$$ 8 \times (f({x+\Delta x}) - f({x-\Delta x})) - (f({x+2\Delta x}) - f({x-2\Delta x})) $$

$$ = 8 \times (2\Delta xf'(x) + \frac{\Delta x^3}{3}f'''(x) + \frac{\Delta x^5}{60}f'''''(x) + ...) - (4\Delta xf'(x) + \frac{8\Delta x^3}{3}f'''(x) + \frac{8\Delta x^5}{15}f'''''(x) + ...) $$

$$ = 16\Delta xf'(x) + \frac{8\Delta x^3}{3}f'''(x) + \frac{2\Delta x^5}{15}f'''''(x) - 4\Delta xf'(x) - \frac{8\Delta x^3}{3}f'''(x) - \frac{8\Delta x^5}{15}f'''''(x) + ... $$

$$ = 12\Delta xf'(x) - \frac{2\Delta x^5}{5}f'''''(x), $$

Rewrite to

$$ -f(x + 2\Delta x) + 8 f(x+\Delta x)- 8 f(x - \Delta x) + f(x - 2\Delta x) $$

$$ = 12\Delta xf'(x) - \frac{2\Delta x^5}{5}f'''''(x) + ..., $$

The first-order derivative is

$$ f'(x) = \frac{-f(x + 2\Delta x) + 8 f(x+\Delta x)- 8 f(x - \Delta x) + f(x - 2\Delta x) + \frac{2\Delta x^5}{5}f'''''(x) + ...}{12 \Delta x}, $$


The fourth-order central difference approximation to the derivative of $f$ at location $x$ is given that

$$ f'(x)\approx \frac{-f(x + 2\Delta x) + 8 f(x+\Delta x)- 8 f(x - \Delta x) + f(x - 2\Delta x)}{12 \Delta x}, $$

The local truncation error is 

$$ \frac{\frac{2\Delta x^5}{5}f'''''(x) + ...}{12 \Delta x} $$

$$ = \frac{\Delta x^4}{30}f'''''(x) + ...$$

The term $\frac{\Delta x^4}{30}f'''''(x)$ shows the major error is proportion to $\Delta x^4$, making it a fourth-order approximation. (Kiusalaas, J. (2014))

## Answer

```python
# define difference approximation
def FirstOrder_diff_approx(f, x, dx):
    """
    Compute the first-order forward difference approximation of the derivative of a function.

    Parameters:
    -----------
    f : callable
        The function to differentiate.
    x : float
        The evaluation point.
    dx : float
        The step size for the finite difference approximation.

    Returns:
    --------
    float
        The approximate derivative of the function at the given point using first-order forward difference.
    """
    return (f(x+dx) - f(x))/dx

def SecondOrder_diff_approx(f, x, dx):
    """
    Compute the second-order central difference approximation of the derivative of a function.

    Parameters:
    -----------
    f : callable
        The function to differentiate.
    x : float
        The evaluation point.
    dx : float
        The step size for the finite difference approximation.

    Returns:
    --------
    float
        The approximate derivative of the function at the given point using second-order central difference.
    """
    return (f(x+dx) - f(x-dx))/(2 * dx)

def FourthOrder_diff_approx(f, x, dx):
    """
    Compute the fourth-order central difference approximation of the derivative of a function.

    Parameters:
    -----------
    f : callable
        The function to differentiate.
    x : float
        The evaluation point.
    dx : float
        The step size for the finite difference approximation.

    Returns:
    --------
    float
        The approximate derivative of the function at the given point using fourth-order central difference.
    """
    return (-f(x+2*dx) + 8 * f(x+dx) - 8 * f(x-dx) + f(x-2*dx))/(12*dx)

def exp_function(x):
    """
    Compute the exponential function e^x.

    Parameters:
    -----------
    x : float
        The input value.

    Returns:
    --------
    float
        The value of the exponential function e^x.
    """
    return np.exp(x)

# implement these approximations
x = 1
dx = 0.00001
diff_1st = FirstOrder_diff_approx(exp_function, x, dx)
diff_2nd = SecondOrder_diff_approx(exp_function, x, dx)
diff_4th = FourthOrder_diff_approx(exp_function, x, dx)
# assert the approximations output same value as actual diff at x = 1
assert np.isclose(diff_1st, np.exp(1))
assert np.isclose(diff_2nd, np.exp(1))
assert np.isclose(diff_4th, np.exp(1))
```

```python
xi = np.linspace(0.9, 1.1, 100)
dxi = 0.001

diff_1st_plot = FirstOrder_diff_approx(exp_function, xi, dxi)
diff_2nd_plot = SecondOrder_diff_approx(exp_function, xi, dxi)
diff_4th_plot = FourthOrder_diff_approx(exp_function, xi, dxi)

# plot
fig = plt.figure(figsize=(18, 6))
ax1 = fig.add_subplot(111)
ax1.plot(xi, diff_1st_plot, label="1st")
ax1.plot(xi, diff_2nd_plot, label="2nd")
ax1.plot(xi, diff_4th_plot, label="4th")
ax1.plot(xi, np.exp(xi), label="1st")
ax1.set_xlabel('$x$', fontsize=16)
ax1.set_ylabel("$f'(x)$", fontsize=16)
ax1.set_title('1st, 2nd, 4th order approx with exact diff', fontsize=16)
ax1.grid(True)
ax1.legend(loc='best', fontsize=8);
```

```python
import numpy as np
import matplotlib.pyplot as plt

def piecewise_interp(xi, yi, num_of_pieces, order, show_im=False):
    """
    Perform piecewise polynomial interpolation on a set of data points.

    The data is divided into 'num_of_pieces' segments, and a polynomial of a given 'order'
    is fitted to each segment. Optionally, the interpolation can be visualized.

    Parameters
    ----------
    xi : array_like
        The x-coordinates of the data points.
    yi : array_like
        The y-coordinates of the data points, corresponding to xi.
    num_of_pieces : int
        The number of pieces the data range is divided into for piecewise interpolation.
    order : int
        The order of the polynomial to fit to each piece.
    show_im : bool, optional
        A flag to indicate if the interpolation should be visualized. Default is False.

    Returns
    -------
    dict
        A dictionary where the keys are the starting x-coordinate of each piece and the values
        are `np.poly1d` objects representing the fitted polynomial for that piece.

    """
    piece_len = len(xi) // num_of_pieces
    x = np.linspace(np.min(xi), np.max(xi), 1000)
    polynomials = {}
    P2 = np.zeros_like(x)

    for i in range(num_of_pieces):
        start = i * piece_len
        # Handle the last piece separately if xi is not exactly divisible by num_of_pieces
        if i == num_of_pieces - 1:
            end = len(xi)
        else:
            end = (i + 1) * piece_len
        
        x_raw = xi[start:end]
        y_raw = yi[start:end]
        coeffs = np.polyfit(x_raw, y_raw, order)
        p = np.poly1d(coeffs)
        polynomials[xi[start]] = p

        # Use np.logical_and to handle interval masking
        x_mask = np.logical_and(x >= xi[start], x < xi[end]) if i < num_of_pieces - 1 else (x >= xi[start])
        P2[x_mask] = p(x[x_mask])

    if show_im:
        plt.figure(figsize=(8, 6))
        plt.plot(xi, yi, 'bo', label='Raw Data')
        plt.plot(x, P2, 'r', label='Piecewise Polynomial')
        plt.title('Raw Data and Piecewise Polynomial Approximations', fontsize=16)
        plt.legend(loc='best', fontsize=14)
        plt.show()

    return polynomials

def piecewise_poly_interp(polynomials, x_test, num_of_pieces):
    """
    Evaluate a piecewise polynomial interpolation at specified test points.

    Given a set of piecewise polynomials and test points, this function evaluates the
    interpolated y-values. It assumes that 'polynomials' are sorted by their keys, and
    these keys represent the start of each piece's interval.

    Parameters
    ----------
    polynomials : dict
        A dictionary of `np.poly1d` objects representing the piecewise polynomials. The keys
        should be the starting x-coordinate of the interval that the corresponding polynomial
        is valid for.
    x_test : array_like
        The x-coordinates where the piecewise polynomial interpolation is evaluated.
    num_of_pieces : int
        The number of pieces the data range is divided into for piecewise interpolation.

    Returns
    -------
    ndarray
        The interpolated y-values at the test points.
    """
    
    intervals = sorted(list(polynomials.keys()))
    
    y_interp = np.zeros_like(x_test)

    for i in range(num_of_pieces):
        
        start = intervals[i]
        
        end = intervals[i + 1] if i < num_of_pieces - 1 else max(x_test)
        
        x_mask = np.logical_and(x_test >= start, x_test < end) if i < num_of_pieces - 1 else (x_test >= start)
        
        y_interp[x_mask] = polynomials[start](x_test[x_mask])
        
    return y_interp
```

```python
# Generate Chebyshev nodes
def chebyshev_nodes(n, a=-1, b=1):
    return 0.5 * (a + b) + 0.5 * (b - a) * np.cos((2 * np.arange(1, n + 1) - 1) / (2 * n) * np.pi)

# Perform polynomial interpolation and plot
def plot_interpolation(true_func, sub_intervals, x_values):
    y_true = true_func(x_values)

    plt.figure(figsize=(12, 8))

    # Plot the true function
    plt.plot(x_values, y_true, 'k', label='True Function')

    for n in sub_intervals:
        # Generate Chebyshev nodes for the current sub-interval
        nodes = chebyshev_nodes(n)
        y_nodes = true_func(nodes)

        # Polynomial interpolation
        coefficients = np.polyfit(nodes, y_nodes, deg=min(n-1, len(nodes)-1))
        p = np.poly1d(coefficients)

        # Evaluate the polynomial at the x_values
        y_interp = p(x_values)

        # Calculate the error
        error = np.abs(y_true - y_interp)

        # Plot the polynomial interpolation
        plt.plot(x_values, y_interp, label=f'Order {min(n-1, len(nodes)-1)} poly, {n} nodes, max error = {max(error):.2e}')
        plt.scatter(nodes, y_nodes, edgecolors='r', facecolors='none', zorder=5)

    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.title('Piecewise Polynomial Interpolation with Non-Uniform Sub-Intervals')
    plt.grid(True)
    plt.show()

# Number of sub-intervals for Chebyshev nodes
sub_intervals = [5, 10, 15]

# Range of x values for plotting
x_values = np.linspace(-1, 1, 1000)

# Plot the interpolation results
plot_interpolation(our_func, sub_intervals, x_values)
```

### Analysis
As the original function varies a lot in the middle of the interval but less away from the origin when we make a piecewise polynomial approximation, we should focus on the middle area, which means allocating more intervals around the area. For example, when we make an approximation away from the origin, we construct a local polynomial fit to data from 4 points simultaneously. However, when we make an approximation near the origin, we can narrow the interval length to, e.g. construct a local polynomial fit to data from 2 points only. This gives the polynomial a higher accuracy with a lower order of polynomials.

It can be concluded that when using uniform-sized sub-intervals, the error decreases as the number of sub-intervals increases. As there are more sub-intervals near x=0 (the gradient sudden change area), it is possible to use lower-order polynomials to approach the accuracy that higher-order polynomials with uniform sub-intervals can do.

```python
def midpoint_rule(a, b, function, number_intervals=10):
    """
    Calculate the numerical integration of a function using the midpoint rule.

    The interval [a, b] is divided into 'number_intervals' equal parts, and the function
    is evaluated at the midpoint of each subinterval. The sum of these evaluations,
    each multiplied by the subinterval width, approximates the integral of the function.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    number_intervals : int, optional
        The number of subintervals to divide [a, b] into (default is 10).

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    interval_size = (b - a)/number_intervals
    xi = [a + i*interval_size for i in range(number_intervals+1)]
    assert interval_size > 0
    assert type(number_intervals) == int
    
    I_M = 0.0
    for i in range(number_intervals):
        I_M += function((xi[i+1]+xi[i])/2)*(xi[i+1] - xi[i])
    return I_M


def trapezoidal_rule(a, b, function, number_intervals=10):
    """
    Calculate the numerical integration of a function using the trapezoidal rule.

    The interval [a, b] is divided into 'number_intervals' equal parts. The function is
    evaluated at each endpoint of these subintervals. Each pair of adjacent evaluations
    forms a trapezoid with the x-axis, and the sum of the areas of these trapezoids
    approximates the integral of the function.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    number_intervals : int, optional
        The number of subintervals to divide [a, b] into (default is 10).

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    interval_size = (b - a)/number_intervals
    xi = [a + i*interval_size for i in range(number_intervals+1)]
    assert interval_size > 0
    assert type(number_intervals) == int
    I_T = 0.0
    for i in range(number_intervals):
        I_T += ((function(xi[i+1]) + function(xi[i]))/2)*(xi[i+1] - xi[i])
        
    return I_T


def simpsons_rule(a, b, function, number_intervals=10):
    """
    Calculate the numerical integration of a function using Simpson's rule.

    The interval [a, b] is divided into 'number_intervals' equal parts. The function is
    evaluated at each endpoint of these subintervals and at the midpoints. Simpson's rule
    applies a weight of 1/6 to the endpoints and 4/6 to the midpoint, multiplying by the
    subinterval width to approximate the integral of the function.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    number_intervals : int, optional
        The number of subintervals to divide [a, b] into (default is 10).

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    Notes
    -----
    The implementation here is different from some others, like SciPy, which may pass
    only discrete data to the function. Here, the function itself is evaluated at the
    necessary points.

    """

    interval_size = (b - a)/number_intervals
    xi = [a + i*interval_size for i in range(number_intervals+1)]
    assert interval_size > 0
    assert type(number_intervals) == int
    
    I_S = 0.0
    for i in range(number_intervals):
        I_S += ((xi[i+1] - xi[i])/6.) * (function(xi[i]) + 4 *
                                         function((xi[i] + xi[i+1])/2.0) + function(xi[i+1]))

    return I_S
```

```python
def improved_trapezoidal_rule(a, b, function, number_intervals=10):
    """
    Calculate the numerical integration of a function using an improved trapezoidal rule,
    while also counting the number of function evaluations.

    This improved version of the trapezoidal rule reduces the number of function evaluations
    by reusing the evaluation from the end of one interval as the start of the next.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    number_intervals : int, optional
        The number of subintervals to divide [a, b] into (default is 10).

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.
    int
        The total number of function evaluations performed, which is optimized by reusing values.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    interval_size = (b - a)/number_intervals
    xi = [a + i*interval_size for i in range(number_intervals+1)]
    # Some examples of some asserts which might be useful here - 
    # you should get into the habit of using these sorts of checks as much as is possible/sensible.
    assert interval_size > 0
    assert type(number_intervals) == int
    count = 0
    I_T = 0.0
    for i in range(number_intervals):
        if i == 0:
            current = function(xi[i])
            count += 1
        else:
            current = current
        next_ = function(xi[i+1])
        count += 1
        I_T += ((next_ + current)/2)*(xi[i+1] - xi[i])
        current = next_
        
    return I_T, count


def improved_simpsons_rule(a, b, function, number_intervals=10):
    """
    Calculate the numerical integration of a function using an improved Simpson's rule,
    while also counting the number of function evaluations.

    This improved version of Simpson's rule reduces the number of function evaluations
    by reusing the evaluation from the end of one interval as the start of the next.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    number_intervals : int, optional
        The number of subintervals to divide [a, b] into (default is 10).

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.
    int
        The total number of function evaluations performed, with two evaluations per interval
        plus one for the initial point.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    interval_size = (b - a)/number_intervals
    xi = [a + i*interval_size for i in range(number_intervals+1)]
    assert interval_size > 0
    assert type(number_intervals) == int
    count = 0
    I_S = 0.0
    for i in range(number_intervals):
        if i == 0:
            current = function(xi[i])
            count += 1
        else:
            current = current
        next_ = function(xi[i+1])
        I_S += ((xi[i+1] - xi[i])/6.) * (current + 4 *
                                         function((xi[i] + xi[i+1])/2.0) + next_)
        count += 2
    return I_S, count
```

```python
def improved_trapezoidal_rule2(a, b, function, xi):
    """
    Calculate the numerical integration of a function using an improved trapezoidal rule,
    while also counting the number of function evaluations.

    This improved version of the trapezoidal rule reduces the number of function evaluations
    by reusing the evaluation from the end of one interval as the start of the next.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    xi : array
        The interval edge.

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.
    int
        The total number of function evaluations performed, which is optimized by reusing values.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    count = 0
    I_T = 0.0
    for i in range(len(xi)-1):
        if i == 0:
            current = function(xi[i])
            count += 1
        else:
            current = current
        next_ = function(xi[i+1])
        count += 1
        I_T += ((next_ + current)/2)*(xi[i+1] - xi[i])
        current = next_
        
    return I_T, count


def improved_simpsons_rule2(a, b, function, xi):
    """
    Calculate the numerical integration of a function using an improved Simpson's rule,
    while also counting the number of function evaluations.

    This improved version of Simpson's rule reduces the number of function evaluations
    by reusing the evaluation from the end of one interval as the start of the next.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    xi : array
        The interval edge.

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.
    int
        The total number of function evaluations performed, with two evaluations per interval
        plus one for the initial point.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    count = 0
    I_S = 0.0
    for i in range(len(xi)-1):
        if i == 0:
            current = function(xi[i])
            count += 1
        else:
            current = current
        next_ = function(xi[i+1])
        I_S += ((xi[i+1] - xi[i])/6.) * (current + 4 *
                                         function((xi[i] + xi[i+1])/2.0) + next_)
        count += 2
    return I_S, count

def midpoint_rule_count2(a, b, function, xi):
    """
    Calculate the numerical integration of a function using the midpoint rule,
    while also counting the number of function evaluations.

    This function is similar to `midpoint_rule` but also returns the total number
    of function evaluations made during the calculation.

    Parameters
    ----------
    a : float
        The lower limit of integration.
    b : float
        The upper limit of integration.
    function : callable
        The function to integrate. Must be a function of a single variable.
    xi : array
        The interval edge.

    Returns
    -------
    float
        The approximate integral of 'function' from 'a' to 'b'.
    int
        The total number of function evaluations performed.

    Raises
    ------
    AssertionError
        If the interval size is not positive or number_intervals is not an integer.

    """
    count = 0
    I_M = 0.0
    for i in range(len(xi)-1):
        I_M += function((xi[i+1]+xi[i])/2)*(xi[i+1] - xi[i])
        count += 1
    return I_M, count
```
